```python
raw_conf = pd.read_csv('confirmed.csv')
raw_death = pd.read_csv('deaths.csv')
raw_rec = pd.read_csv('recovered.csv')
```

```python
raw_conf.shape
```

```python
raw_death.shape
```

```python
raw_rec.shape
```

```python
raw_rec
```

```python
raw_conf2 = pd.melt(raw_conf,id_vars=['Province/State','Country/Region','Lat','Long'],
                    var_name=['Date'])
```

```python
raw_conf2
```

```python
raw_death2 = pd.melt(raw_death,id_vars=['Province/State','Country/Region','Lat','Long'],
                    var_name=['Date'])
```

```python
raw_rec2 = pd.melt(raw_rec,id_vars=['Province/State','Country/Region','Lat','Long'],
                    var_name=['Date'])
```

```python
raw_death2
```

```python
raw_rec2
```

```python
## Transforming to datetime object
raw_conf2['Date'] = pd.to_datetime(raw_conf2['Date'])
raw_death2['Date'] = pd.to_datetime(raw_death2['Date'])
raw_rec2['Date'] = pd.to_datetime(raw_rec2['Date'])
```

```python
## Renaming value to proper name
raw_conf2.columns = raw_conf2.columns.str.replace('value','Confirmed')
raw_death2.columns = raw_death2.columns.str.replace('value','Deaths')
raw_rec2.columns = raw_rec2.columns.str.replace('value','Recovered')
```

```python
## Dealing with Null values
raw_conf2.isnull().sum()
```

```python
raw_conf2['Province/State'].fillna(raw_conf2['Country/Region'],inplace=True)
raw_death2['Province/State'].fillna(raw_death2['Country/Region'],inplace=True)
raw_rec2['Province/State'].fillna(raw_rec2['Country/Region'],inplace=True)
```

```python
raw_conf2.isnull().sum() ## Now there is no Null values present in df
```

```python
## JOIN OPERATION
```

```python
full_join = raw_conf2.merge(raw_death2[['Province/State','Country/Region','Date','Deaths']],
                           how='left',
                           left_on=['Province/State','Country/Region','Date'],
                           right_on=['Province/State','Country/Region','Date'])
```

```python
full_join
```

```python
full_join = full_join.merge(raw_rec2[['Province/State','Country/Region','Date','Recovered']],
                           how='left',
                           left_on=['Province/State','Country/Region','Date'],
                           right_on=['Province/State','Country/Region','Date'])
```

```python
full_join
```

```python
full_join
```

```python
# Adding a new column
full_join['Month-Year'] = full_join['Date'].dt.strftime('%m-%Y')
```

```python
full_join
```
